```python
# Better to change the columns name for non-error detect
scp.columns = ["CarMake", "CarModel", "Year", "EngineSize(L)", "HorsePower", "Torque(lb-ft)", "0-60_MPH(seconds)",
                                                                                              "Price(in USD)"]
```

```python
scp.shape
```

**Missings and Converting process is needed to start DA**

```python
# Replacing unWanted Char's from the set. Also removing missings 

scp.replace({',': ''}, regex=True, inplace =True) # replacing all unwanted character from the whole set
scp.dropna(inplace = True) #drop all missing.
```

```python
scp = scp[scp["HorsePower"].str.contains("\+", regex=True) == False] # need to remove 

#changing the Dtypes of the columns.
scp['HorsePower'] = scp['HorsePower'].astype(float) 

scp['Torque(lb-ft)'] = scp['Torque(lb-ft)'].astype(float)

scp['0-60_MPH(seconds)'] = scp['0-60_MPH(seconds)'].astype(float)

scp["Price(in USD)"] = scp["Price(in USD)"].astype(float)

scp['CarType'] = np.where((scp['CarMake'] == 'Tesla') | (scp['EngineSize(L)'].str.contains('Electric')),
                                                                                    'Electric', 'Gas')
```

### **Outliers counting**

```python
temp = scp.copy()
```

```python
def outlier_detect(columns):
    for column in columns: 
        for x in [column]:
            q75,q25 = np.percentile(temp.loc[:,x],[75,25])
            intr_qr = q75-q25

            max = q75+(1.5*intr_qr)
            min = q25-(1.5*intr_qr)

            temp.loc[temp[x] < min,x] = np.nan
            temp.loc[temp[x] > max,x] = np.nan
```

```python
numeric_data = scp.select_dtypes(include=[np.number]).columns
```

```python
outlier_detect(list(numeric_data))
```
